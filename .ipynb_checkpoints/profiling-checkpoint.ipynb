```python
import glets.raytrace as rt
import glets.gbd as gf
import glets.utils as ut
import numpy as np
import matplotlib.pyplot as plt
#!pip install poppy==1.0.0
#!pip install --upgrade pip # import poppy
#!pip install git+https://github.com/spacetelescope/poppy
import poppy
# print(poppy.__version__)
import sys
# sys.version
# !pip install numba
from scipy.ndimage import zoom
import astropy.units as u
import time
poppy.conf.use_numexpr = True
poppy.accel_math._USE_NUMEXPR
def compute_mean_std(data):
    return np.mean(data),np.std(data)
```

```python
# Set up poppy osys
nrays_across = 110
scale = 5e-1
detector_size = 0.0014079983 #5e-4
wlen = 2.2e-6
npix = 512
size = 25.4e-3

import astropy.units as u

t1 = time.perf_counter()
wf = poppy.FresnelWavefront(size/2*u.m,wavelength=wlen,npix=npix,oversample=8)
circ = poppy.CircularAperture(radius=size/2*u.m)
zwfe = poppy.ZernikeWFE(coefficients=[0,0,0,2.2e-7*0],radius=size*u.m)
pwfe = poppy.StatisticalPSDWFE(index=-3.0, wfe=100e-9*u.m, radius=size*u.m)
thin = poppy.QuadraticLens(10*size*u.m)

wf *= circ
wf *= zwfe
#wf *= pwfe
opd_to_pass = ut.grab_center(zwfe.get_opd(wf),npix/2)
wf *= thin
z = 10*size*u.m
wf.propagate_fresnel(z)

t2 = time.perf_counter()
print('Fresnel Propagation Time = ',t2-t1)

# plt.figure(figsize=(12,5))
# wf.display('both',colorbar=True,imagecrop=detector_size,scale='log')
# plt.suptitle("After propagating {}".format(z), fontsize=18)

t1 = time.perf_counter()
rays = rt.makerays(25.4e-3, nrays_across,circle=False)
# Give rays some wavefront error
array = opd_to_pass#zernike_nm(3,-3,256)*1e-7 # kolphase(npix)*5e-9 # 
array_gbd = zoom(array,nrays_across/npix)
testrays = rt.ArbitraryWFE(nrays_across, size,rays,scale,array=array_gbd) # array assumes units of meters
nrays_total = int((rays.shape[1]))

t3 = time.perf_counter()
# Set up a thin lens that goes to focus
lens = rt.ThinLens(254e-3, nrays_total)
dist = rt.FreeSpace(254e-3, nrays_total)
rayt = rt.matmultlist(lens, dist)
rayt = rt.matmultlist(dist,rayt)
t4 = time.perf_counter()
print('time to compute rays = ',t4-t3)

t5 = time.perf_counter()
raysincirc = testrays[:,np.sqrt(testrays[0,:]**2 + testrays[1,:]**2) < 12.7e-3]
rtmincirc  = rayt[:,:,np.sqrt(testrays[0,:]**2 + testrays[1,:]**2) < 12.7e-3]
t6 = time.perf_counter()
print('time to index rays = ',t6-t5)

t7 = time.perf_counter()
efie_numerical = gf.eval_gausfield(raysincirc, rtmincirc, wlen, detector_size, npix, use_numexpr=True)
t8 = time.perf_counter()
print('time to compute full gfield = ',t8-t7)

t9 = time.perf_counter()
efie_numerical = np.reshape(efie_numerical,[npix,npix])
t10 = time.perf_counter()
print('time to reshape gfield = ',t10-t9)

t2 = time.perf_counter()
print('GBD Propagation Time = ',t2-t1)
#ut.amp_phase_plot(efie_numerical,logintensity=True)
```

```python
fresnel_np = [6.859439500141889,6.86651138169691,6.850619695149362,6.9468078459613025,6.834841417148709]
gbd_np     = [17.307950691785663,17.906354452949017,17.46287792129442,17.4801528127864,17.5271875471808]
fresnel_ne = [5.936061820015311,6.006530724000186,6.179728501942009,5.830662473104894,6.078931807074696]
gbd_ne     = [6.40507240081206,6.496410802938044,6.385508903302252,6.296534692402929,6.282955896109343]

gbd_np_mean, gbd_np_std = compute_mean_std(gbd_np)
gbd_ne_mean, gbd_ne_std = compute_mean_std(gbd_ne)

fnl_np_mean, fnl_np_std = compute_mean_std(fresnel_np)
fnl_ne_mean, fnl_ne_std = compute_mean_std(fresnel_ne)

labels = ['Numpy','Numexpr']

x = np.linspace(1,2,2)
fig,ax = plt.subplots(figsize=[7,7])
ax.bar(x[0]-.2,gbd_np_mean,yerr=gbd_np_std,color='orange',width=0.4)
ax.bar(x[0]+.2,fnl_np_mean,yerr=fnl_np_std,color='blue',width=0.4)
ax.bar(x[1]-.2,gbd_ne_mean,yerr=gbd_ne_std,color='orange',width=0.4)
ax.bar(x[1]+.2,fnl_ne_mean,yerr=fnl_ne_std,color='blue',width=0.4)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.yaxis.grid(True)
ax.set_title('1876 Beamlets, npix = {}'.format(npix))
ax.set_xlabel('Exponential Calculation Method')
ax.set_ylabel('Total Runtime [s]')
plt.legend(['GBD','Fresnel'])

x = np.linspace(1,2,2)
fig,ax = plt.subplots(figsize=[7,7])
ax.bar(x[0]-.2,gbd_np_mean/gbd_np_mean,color='orange',width=0.4)
ax.bar(x[0]+.2,fnl_np_mean/fnl_np_mean,color='blue',width=0.4)
ax.bar(x[1]-.2,gbd_np_mean/gbd_ne_mean,color='orange',width=0.4)
ax.bar(x[1]+.2,fnl_np_mean/fnl_ne_mean,color='blue',width=0.4)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.yaxis.grid(True)
ax.set_title('1876 Beamlets, npix = {}'.format(npix))
ax.set_xlabel('Exponential Calculation Method')
ax.set_ylabel('Relative Speedup to Numpy')
plt.legend(['GBD','Fresnel'])
```

```python
from matplotlib.colors import LogNorm
from scipy.ndimage import shift
print(raysincirc.shape)

# fix shift in beamlet calculation
efie_num_shift = shift(np.abs(efie_numerical),[0,0])

efresnel_pars = np.flip(ut.grab_center(wf.wavefront,cut=64),axis=0)
efresnel_pars = np.abs(efresnel_pars)
#efresnel_pars = np.abs(efresnel_pars*np.conj(efresnel_pars))
efresnel_pars /= np.sum(efresnel_pars)
efresnel_pars /= np.max(efresnel_pars)

enumeric_pars = ut.grab_center(efie_num_shift,cut=64)
enumeric_pars = np.abs(enumeric_pars)
#enumeric_pars = np.abs(enumeric_pars*np.conj(enumeric_pars))
enumeric_pars /= np.sum(enumeric_pars)
enumeric_pars /= np.max(enumeric_pars)

efresnel_pars = shift(efresnel_pars,[.5,-.5])
ediff = np.abs(enumeric_pars - efresnel_pars)#/efresnel_pars*100
ediff[ediff > 1e1] = False
def array2poppyhdul(array,pixscal):
    
    from astropy.io import fits
    phdu = fits.PrimaryHDU(array)
    hdul = fits.HDUList([phdu])
    hdul[0].header['PIXELSCL'] = pixscal
    
    return hdul

fnl_pro = poppy.radial_profile(array2poppyhdul(efresnel_pars,2.74999666e-06))
gbd_pro = poppy.radial_profile(array2poppyhdul(enumeric_pars,2.74999666e-06))
dif_pro = poppy.radial_profile(array2poppyhdul(ediff,2.74999666e-06))

vmin = 1e-4
vmax = 1

print('StdDev of Difference = ',np.ma.std(ediff))
print('Mean of Difference = ',np.ma.mean(ediff))
plt.figure(figsize=[20,7])
plt.set_cmap('plasma')
plt.suptitle('Peak-Normalized Point-Spread Functions (PSF)')
plt.subplot(131)
plt.title('GBD PSF')
plt.imshow(enumeric_pars,norm=LogNorm(),vmin=vmin,vmax=vmax)
plt.colorbar()
plt.subplot(132)
plt.title('Fresnel PSF')
plt.imshow(efresnel_pars,norm=LogNorm(),vmin=vmin,vmax=vmax)
plt.colorbar()
plt.subplot(133)
plt.title('Difference')
plt.imshow(ediff,vmin=vmin,vmax=vmax,norm=LogNorm())
plt.colorbar()
plt.show()

eye = np.identity(128)

gbd_m45 = enumeric_pars[eye == 1]
fnl_m45 = efresnel_pars[eye == 1]
gbd_hrz = enumeric_pars[int(enumeric_pars.shape[0]/2),:]
fnl_hrz = efresnel_pars[int(efresnel_pars.shape[0]/2),:]

gbd_m45 = gbd_m45[int(gbd_m45.size/2):]
gbd_hrz = gbd_hrz[int(gbd_hrz.size/2):]
fnl_m45 = fnl_m45[int(fnl_m45.size/2):]
fnl_hrz = fnl_hrz[int(fnl_hrz.size/2):]

plt.figure(figsize=[14,7])
#plt.plot(fnl_hrz,label='Fresnel 0$^o$',color='black')
#plt.plot(fnl_m45,label='Fresnel 45$^o$',color='red')
#plt.plot(gbd_hrz,label='GBD 0$^o$',color='black',linestyle='solid',linewidth=4,alpha=.4)
#plt.plot(gbd_m45,label='GBD 45$^o$',color='red',linestyle='solid',linewidth=4,alpha=.4)
plt.plot(fnl_pro[1],label='Fresnel Radial Average',color='black',linewidth=3,linestyle='solid')
plt.plot(gbd_pro[1],label='GBD Radial Average',color='red',linewidth=3,linestyle='dashdot')
plt.plot(dif_pro[1],label='Difference Radial Average',color='black',linewidth=3,linestyle='dotted')
plt.yscale('log')
plt.xlim([0,60])
plt.xlabel('Radial Distance [$\lambda/D$]')
plt.ylabel('Peak-Normalized Intensity')
#plt.text(6,6,'$\sigma_d=${}'.format(np.std(ediff)))
plt.xticks(ticks=[0,12,24,36,48],labels=[0,1.5,3,4.5,6])
plt.legend()
plt.show()
```
